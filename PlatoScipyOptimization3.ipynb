```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from scipy import interpolate
import pandas as pd
```

```python
url="https://raw.githubusercontent.com/pedrogonzalezbeermann/FittingModels/master/MedioBasal.csv"
df1 = pd.read_csv(url)
df1
```

```python
def f(p):
   M=p[1]+p[2]/(1+np.exp(p[4]*(x2-p[3])))
   yy=(p[0]*y2-p[0]/M*y2*y2)*dt
   s=np.cumsum(yy)+1.0
   
   return np.sum((y2-s)**2)
```

```python

columna=df1.columns
n=len(columna)
mu=[]
for plato in range(1,n):

    x=df1["dia"]
    y=df1[df1.columns[plato]]

    xmax=np.max(x)

    g=interpolate.interp1d(x,y,kind='cubic')  #zero, cubic, slinear, quadratic
    dt=0.01
    x2=np.arange(0,xmax,dt)
    y2=g(x2)



    x0=np.array([1.59,1.27,48.7,1.87,-41.87])
    res = minimize(f,x0, method='Nelder-Mead',options={'xtol': 1e-8, 'disp': True,'maxiter': 15000})

    ##Para graficar
    p=np.array(res.x)
    M=p[1]+p[2]/(1+np.exp(p[4]*(x2-p[3])))
    yy=(p[0]*y2-p[0]/M*y2*y2)*dt
    s=np.cumsum(yy)+1.0
    plt.figure(figsize=(8,8))
    plt.scatter(x,y)
    plt.plot(x2,s)
    
    plt.text(1,20,r'$\mu=$'+str(p[0]),fontsize=14)
    plt.xlabel("t(dias)")
    plt.ylabel("Area")

    plt.title(columna[plato],fontsize=20)
    plt.show() 
    mu.append(p[0])
```

```python
d = {'Tratamieno': columna[1:], 'mu': mu}
df = pd.DataFrame(data=d)
df
```
